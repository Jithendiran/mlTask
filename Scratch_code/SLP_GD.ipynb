```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5]) +.0
y = np.array([2,4,6,8,10]) +.0
w = np.random.rand()
b = np.random.rand()
lr = 0.01
for i in range(20):
    y_pred = (x*w) + b
    err = (y_pred - y) 
    w -= lr * ((x * err).sum() / len(x) )
    b -= lr * ((err).sum() / len(x) )
    print(f'Epoch {i}, pred: {y_pred[1]-y_pred[0]}, Error {(err ** 2).mean()}')
```

```python
m = 100  # Number of samples
n = 2    # Number of features
X = np.random.rand(m, n)
y = 3 * X[:, 0] + 5 * X[:, 1] + 2 + 0.1 * np.random.randn(m)  # True coefficients: [3, 5], Intercept: 2
x1 = X[:,0]
x2 = X[:,1]
#chat gpt data set 
```

```python
lr = 0.01
b = np.random.rand()
w1 = np.random.rand()
w2 = np.random.rand()

# Gradient Descent or Batch Gradient Descent or Vanilla Gradient Descent
for epoch in range(20):
    y_p = ((x1* w1) + (x2 * w2)) +  b

    err = y_p - y

    w1 -= lr * ((x1 * err).sum() / len(x1)) 
    w2 -= lr * ((x2 * err).sum() / len(x2)) 
    b -= lr * (err.sum() / len(err))

    print(f'Epoch {epoch}, Cost: {np.mean(err ** 2)}')
```

```python
# Stochastic Gradient Descent
b = 0.2#np.random.rand()
w1 = 0.043#np.random.rand()
w2 = 0.612#np.random.rand()

# calculate gradient for 1 sample
for epoch in range(20):

    y_p = ((x1* w1) + (x2 * w2)) +  b

    ind = np.random.choice(X.shape[0],1,replace=False)[0]
    err = y_p[ind] - y[ind]

    w1 -= lr * ((x1[ind] * err)) 
    w2 -= lr * ((x2[ind] * err))

    b -= lr * (err)

    print(f'Epoch {epoch}, Cost: {np.mean(err ** 2)}')
```

```python
# mini batch
batch = 12
b = 0.2#np.random.rand()
w1 = 0.043#np.random.rand()
w2 = 0.612#np.random.rand()

# calculate gradient for random subset of array
for epoch in range(20):

    y_p = ((x1* w1) + (x2 * w2)) +  b

    ind = np.random.choice(X.shape[0], batch,replace=False)
    
    err = y_p[ind] - y[ind]

    w1 -= lr * ((x1[ind] * err).sum() / batch) 
    w2 -= lr * ((x2[ind] * err).sum() / batch)

    b -= lr * (err.sum() / batch)

    print(f'Epoch {epoch}, Cost: {np.mean(err ** 2)}')
```

```python
# momntum with stochastic Gradient Descent
# Method 1
vw1,vw2,vb = 0,0,0 # initial value for momentum
momentum_hyperparameter = 0.98

b = 0.2#np.random.rand()
w1 = 0.043#np.random.rand()
w2 = 0.612#np.random.rand()

# calculate gradient for 1 sample
for epoch in range(20):

    y_p = ((x1* w1) + (x2 * w2)) +  b

    ind = np.random.choice(X.shape[0],1,replace=False)[0]

    err = y_p[ind] - y[ind]

    vw1 = (momentum_hyperparameter * vw1) + ((1 - momentum_hyperparameter) * ((x1[ind] * err)))
    vw2 = (momentum_hyperparameter * vw2) + ((1 - momentum_hyperparameter) * ((x2[ind] * err)))

    vb = (momentum_hyperparameter * vb) + ((1 - momentum_hyperparameter) * (err))

    w1 -= lr * vw1
    w2 -= lr * vw2

    b -= lr * vb

    print(f'Epoch {epoch}, Cost: {np.mean(err ** 2)}')
```

```python
# Nesterov momntum with mini batch

vw1,vw2,vb = 0,0,0 # initial value for momentum
momentum_hyperparameter = 0.98

batch = 12
b = 0.2#np.random.rand()
w1 = 0.043#np.random.rand()
w2 = 0.612#np.random.rand()

# calculate gradient for random subset of array
for epoch in range(20):

    w1 = w1 - (momentum_hyperparameter * vw1)
    w2 = w2 - (momentum_hyperparameter * vw2)
    b = b - (momentum_hyperparameter * vb)


    y_p = ((x1* w1) + (x2 * w2)) +  b

    ind = np.random.choice(X.shape[0], batch,replace=False)
    
    err = y_p[ind] - y[ind]

    vw1 = (momentum_hyperparameter * vw1) + (lr * ((x1[ind] * err).sum() / batch)) 
    vw2 = (momentum_hyperparameter * vw2) + (lr * ((x2[ind] * err).sum() / batch))

    vb = (momentum_hyperparameter * vb) + (lr * (err.sum() / batch))

    w1 -= vw1
    w2 -= vw2

    b -= vb

    print(f'Epoch {epoch}, Cost: {np.mean(err ** 2)}')
```
