# Slope
![img](./assest/s.png) 

 A slope of a line is the change in y coordinate with respect to the change in x coordinate. In the above image blue line is called as slope

$$
  slope = \frac{(y2-y1)}{(x2 - x1)}
$$

Slope of the linear function is constant in any points and Slope of the non-linear function is vary for point to point

## Linear

![img](./assest/p.jpg)

Eg : 1
$$
(x1,y1) = (1,1) \\
(x2,y2) = (2,2) \\
slope = \frac{(y2-y1)}{(x2 - x1)} = \frac{(2-1)}{(2 - 1)} = \frac{1}{1} = 1\\
$$

Eg : 2
$$
(x1,y1) = (1,1) \\
(x2,y2) = (3,3) \\
slope = \frac{(y2-y1)}{(x2 - x1)} = \frac{(3-1)}{(3 - 1)} = \frac{2}{2} = 1\\
$$

## Non-Linear
![img](./assest/x.jpg)

Eg : 1
$$
(x1,y1) = (1,3) \\
(x2,y2) = (2,1) \\
slope = \frac{(y2-y1)}{(x2 - x1)} = \frac{(1-3)}{(2 - 1)} = \frac{-2}{1} = -2\\
$$

Eg : 2
$$
(x1,y1) = (1,3) \\
(x2,y2) = (1,2) \\
slope = \frac{(y2-y1)}{(x2 - x1)} = \frac{(2-3)}{(1 - 1)} = \frac{-1}{0} = \infty \\
$$

Slope values can be 0, +ve, -ve and $\infty$

if x is move by h distance from x then input will be $x+h$ and output will be $(x+h)^2$  
$slope = \frac{(y_2 - y_1)}{(x_2 - x_1)}\\$
$$
\begin{aligned}
x_1 = x\\
y_1 = x^2\\
x_2 = (x+h)\\
y_2 = (x+h)^2 \\

substitute \ the \ values \ to \ the \ equation \\
\\

slope = \frac{((x+h)^2 - x^2)}{( (x + h) - x)} \\
\\
= \frac{((x+h)^2 - x^2)}{( x + h - x)} \\

\\
`(x+h)^2 is\ similar\ to\ (a+b)^2`\\
``(a+b)^2 = a^2 + 2ab + b^2``\\
`From\ formula\ (x+h)^2 = x^2 + 2xh + h^2`\\
\\
= \frac{( x^2 + 2xh + h^2 - x^2)}{( x + h - x)} \\
\\
= \frac{( \cancel{x^2} + 2xh + h^2 \cancel{-x^2})}{( \cancel{x} + h \cancel{-x})} \\

\\
= \frac{ 2xh + h^2 }{h} \\

\\
Divide\ by\ \frac{h}{h} \\
\\
= 2x + h \\
\end{aligned}
$$
In the derivation the result is *$2x+h$*. Here the $h$ is very small so it can be negligible. So final answer is $2x$

when input is 5 it's output is 25 and it's slope is $2*x$ which is $2 * 5 =10$

This is the proof of how $\frac{\partial }{\partial x} = x^2 = 2x$

### Gradient

Finding minimum of slope for this kind of equation is simpler but if function is complex and having multiple x variable is harder to equate the slope

$y = \frac{cos x+e^x}{sin x}$

For the above equation it is harder to get the slope in equation form, for this limitation gradient is used

### Gradient is used to get the maximum of the slope, It is calculated by partial dervative of (x,y).

Gradient is operating on vectors.

$$
\nabla f(x,y) = \begin{bmatrix}
f_x(x,y)\\
f_y(x,y)\\
\end{bmatrix}
$$

Example:
 $$
 f(x,y) = x^2 + y^2 \\
 \frac{\partial}{\partial x} = 2x \\
 \frac{\partial}{\partial y} = 2y \\
\nabla f(x,y) = \begin{bmatrix}
2x\\
2y\\
\end{bmatrix}\\

This\ derivation\ is\ used\ to\ find\ the\ directional\ derivative \\
D_u f(x,y) = \nabla f(x,y) . \overrightarrow{U} \\
\overrightarrow{U} denotes\ unit\ vector(x\, y)points\\

D_u f(x,y) = \nabla \begin{bmatrix}
 \frac{\partial}{\partial x}\\
\frac{\partial}{\partial y}\\
\end{bmatrix}  
*
\begin{bmatrix}
x\\
y\\
\end{bmatrix}\\  
 $$

#### Gradient 
$$  
    x = old_x + \alpha * \nabla x \\
    y = old_y + \alpha * \nabla y \\

    \alpha\ is\ a\ control\ parameter\\
$$

### calculate gradient for previous example
$y=x^2$  
$\frac{\partial y}{\partial x} = 2x$

```python
def grad(x,alpha=0.01):
    '''
        x : the  value
        alpha: control vparameter (0 to 1)
    '''
    return x + (alpha * (((2 * x ) - 2 )+ 0.00001)) #0.00001 is to escape from 0 multiply
```

```python
# consider x is take from 1
x1 = 1
y1 = ((x1 - 1) ** 2) + 1

plt.plot(x,y)
x1 = grad(x1)
y1 = ((x1 - 1) ** 2) + 1 # y value will be change based on x value
plt.scatter(x1,y1)
print(x1,y1)
```

### from the above graph x and y value is increase, do another iteration

```python
plt.plot(x,y)
x1 = grad(x1)
y1 = ((x1 - 1) ** 2) + 1 # y value will be change based on x value
plt.scatter(x1,y1)
print(x1,y1)
```

### Again it is increasing, do it in iteration

```python
plt.plot(x,y)
for i in range(50):
    x1 = grad(x1)
    y1 = ((x1 - 1) ** 2) + 1 # y value will be change based on x value
    plt.scatter(x1,y1)
    print(x1,y1)

plt.show()
```

### It is clearly visible the points are moving in upwords

### Above equation is diffifult to see the action so use $y=x^2$ as a equation it's derivation $\frac{\partial y}{\partial x} = 2x$

```python
def fun(x):
    '''
    x : value
    '''
    return x**2

def grad_fun(x,alpha=0.01):
    '''
    x : value
    lr:  control parameter (0 to 1)
    '''
    return x + (alpha * (2 * x + 0.00001))
```

```python
x = range(-10,11,1)
y = [fun(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
x1 = 1 ## start 1
y1 = fun(x1) ## 1

plt.plot(x,y)
for i in range(10):
    x1 = grad_fun(x1,0.1)
    y1 = fun(x1) # y value will be change based on x value
    plt.scatter(x1,y1)
plt.show()
```

## The points are starting from low and it's like uphilling 
## This gradient is known as gradient asscent
## Downhilling algorithm is known as gradient descent
$$
Gradient\ Descent
x = old_x - \alpha * \nabla x
$$

```python
def grad_fun_desc(x,alpha=0.01):
    '''
    x : value
    lr:  control parameter (0 to 1)
    '''
    return x - (alpha * (2 * x + 0.00001))
```

```python
x1 = 10 ## start 10
y1 = fun(x1) ## 100

plt.plot(x,y)
for i in range(10):
    x1 = grad_fun_desc(x1,0.1)
    y1 = fun(x1)
    plt.scatter(x1,y1)
plt.show()
```

### Now the ponints are downhilling
